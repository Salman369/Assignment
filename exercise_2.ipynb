```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
# Open and read the text file
with open('/content/JaneEyre.txt', 'r') as file:
    text = file.read()

```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

# Generate the word cloud
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color="white",
    stopwords=set(STOPWORDS),
    min_font_size=10
).generate(text)

# Display the word cloud
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

```python
from PIL import Image
import numpy as np

# Load the mask image and convert it to a NumPy array
mask = np.array(Image.open('/content/mask_image.jpg'))
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

# Create the WordCloud object with the mask
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color="white",
    stopwords=set(STOPWORDS),
    mask=mask,  # Apply the mask here
    contour_width=1,
    contour_color='black'
).generate(text)

# Display the word cloud
plt.figure(figsize=(10, 10), facecolor=None)
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
